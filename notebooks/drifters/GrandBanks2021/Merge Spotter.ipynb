A notebook is probably overkill for this...

Spotter data can only be downloaded 2 months at a time. So this notebook merges files into one.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f1='data/SPOT-1017_2021-12-06_2022-02-04_download.csv'
f2='data/SPOT-1017_2022-02-02_2022-02-12_download.csv'

fall='data/SPOT-1017_2021-12-06_2022-02-12_download.csv'
```

```python
d1 = pd.read_csv(f1)
d2 = pd.read_csv(f2)
```

```python
fig,ax=plt.subplots(1,1,figsize=(5,5))
d1.plot(x='Longitude (deg)', y='Latitude (deg)',ax=ax)
d2.plot(x='Longitude (deg)', y='Latitude (deg)',ax=ax)
```

```python
dnew=pd.concat([d2,d1], ignore_index=True)
dnew
```

```python
dnew.drop_duplicates(inplace=True, ignore_index=True, subset='Epoch Time')
```

```python
fig,ax=plt.subplots(1,1,figsize=(5,5))
dnew.plot(x='Longitude (deg)', y='Latitude (deg)',ax=ax)
```
