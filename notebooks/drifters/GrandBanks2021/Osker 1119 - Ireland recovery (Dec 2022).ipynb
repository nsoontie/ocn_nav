```python
def plot_map(ax,lon_min=lon_min, lon_max=lon_max, lat_min=lat_min, lat_max=lat_max):
    bmap = Basemap(projection='merc',llcrnrlon=lon_min, urcrnrlon=lon_max,
                   llcrnrlat=lat_min,urcrnrlat=lat_max, ax=ax,resolution='i')
    bmap.drawcoastlines()
    bmap.fillcontinents(color='grey',)
    bmap.drawparallels(np.arange(lat_min, lat_max, 5.),labels=[1,0,0,0],fontsize=16)
    bmap.drawmeridians(np.arange(lon_min, lon_max,5.),labels=[0,0,0,1],fontsize=16)
    return bmap
```

```python
def plot_bathymetry(bmap, ax, etopo):
    lons = etopo.x.values
    lats = etopo.y.values
    lons, lats = np.meshgrid(lons,lats)
    x, y = bmap(lons, lats)
    bathy = etopo.z.values
    levels = [-500, -100,]
    CS = bmap.contour(x,y,bathy,levels, linestyles='solid', linewidth=0.1,colors='black')
    ax.clabel(CS, CS.levels, inline=True, fmt='%1.0f', fontsize=12)
```

```python
def plot_spotter(fname, bmap):
    d = pd.read_csv(fname)
    x, y = bmap(d['Longitude (deg)'].values,d['Latitude (deg)'].values)
    bmap.plot(x,y,color='b', label='SPOT-1017')
        
```

```python
def plot_oskers(file, bmap, idplot=None):
    d = pd.read_csv(file)
    dg = d.groupby('Device')
    for name, g in dg:
        if idplot:
            if name==idplot:
                x, y = bmap(g['Longitude'].values,g['Latitude'].values)
                bmap.plot(x,y,color='m',lw=5, label='Osker 1119')
                print(name)
```

```python
fig, ax = plt.subplots(1,1,figsize=(15,10))
bmap = plot_map(ax)
#plot_bathymetry(bmap,ax,esel)
plot_spotter(spotter, bmap)
plot_oskers(oskers, bmap, idplot=osker_id)
x,y=bmap(-52.7126,47.5615)
ax.plot(x,y,'g*',ms=20,label="St. John's, Newfondland, Canada")
x,y=bmap(lon_recovery, lat_recover)
ax.plot(x,y,'rs',ms=10,label="Approximate recovery location of Osker 1119")
ax.legend(fontsize=14)
```
